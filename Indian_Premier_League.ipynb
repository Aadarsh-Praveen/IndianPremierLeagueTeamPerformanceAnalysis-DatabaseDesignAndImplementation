# Query 1

```python
sql_select_Query = "select * from player limit 20"
cursor = connection.cursor()
cursor.execute(sql_select_Query)
records = cursor.fetchall()
print("Player_Details:\n")
for row in records:
    print(row,"\n")
```

# Query 2

```python
sql_select_Query = "select count(Player_ID) as Players from Batter where Runs > 10000;"
cursor = connection.cursor()
cursor.execute(sql_select_Query)
records = cursor.fetchall()
print("Players with more than 10000 runs:\n")
for row in records:
    print(row[0],"\n")
```

# Query 3

```python
sql_select_Query = "select (P.FirstName) as Name, P.Nationality, B.Runs, B.High_Score, Bo.Wickets from Player P join Batter B on P.Player_ID = B.Player_ID join Bowler Bo on P.Player_ID = Bo.Player_ID order by P.Nationality limit 20;"
cursor = connection.cursor()
cursor.execute(sql_select_Query)
records = cursor.fetchall()
print("Player with their total runs and high score and their total wickets:\n")
for row in records:
    print(row,"\n")
```

# Query 4

```python
sql_select_Query = "Select * from Batter where Player_id in (select Player_ID from Player where Player_Type = 'All Rounder') limit 20;"
cursor = connection.cursor()
cursor.execute(sql_select_Query)
records = cursor.fetchall()
print("Players who are all rounders:\n")
for row in records:
    print(row,"\n")
```

# Query 5

```python
sql_select_Query = "Select Player_ID, FirstName, LastName, Nationality from Player where Nationality = 'New Zealand' UNION Select Player_ID, FirstName, LastName, Nationality from Player where Nationality = 'Zimbabwe';"
cursor = connection.cursor()
cursor.execute(sql_select_Query)
records = cursor.fetchall()
print("Players who are from 'New Zealand' and 'Zimbabwe':\n")
for row in records:
    print(row,"\n")
```

# Query 6

```python
sql_select_Query = "SELECT * FROM Player as p WHERE NOT EXISTS (SELECT 1 FROM Player as p1 WHERE Player_Type = 'All Rounder' AND p1.Player_ID = p.Player_ID) limit 20;"
cursor = connection.cursor()
cursor.execute(sql_select_Query)
records = cursor.fetchall()
print("Players who are Bowlers and Batsman:\n")
for row in records:
    print(row,"\n")
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
#Line plot for No of runs
```

```python
query = """select runs from batter;"""""
query1_df = pd.read_sql(query, connection)
plt.figure(figsize = (8,4))
sns.scatterplot(query1_df.runs, color = "red")
plt.title("Runs Scored by Batsman")
plt.ylabel("Runs")
plt.tight_layout()
plt.show()
```

```python
#Barplot to calculate the mentor experience in years
```

```python
query = """select Mentor_Name, Experience from mentor;"""""
query2_df = pd.read_sql(query, connection)
plt.figure(figsize = (8,4))
plt.barh(query2_df.Mentor_Name, query2_df.Experience, color = "skyblue")
plt.title("MENTOR EXPERIENCE IN YEARS")
plt.ylabel("Mentor Name")
plt.xlabel("Years")
plt.tight_layout()
plt.show()
```
